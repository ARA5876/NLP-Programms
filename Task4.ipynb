```python
from urllib import request
```

```python
url = "https://www.gutenberg.org/files/69965/69965-0.txt"
```

```python
response= request.urlopen(url)
```

```python
raw = response.read().decode('utf8')
```

```python
import nltk

from nltk.tokenize import word_tokenize
tokens = word_tokenize(raw)
```

```python
print(tokens[:23])
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
#partsofspeechtagging
tagged=nltk.pos_tag(tokens)
```

```python
print(tagged)
```

#### Stemming 

```python
from nltk.stem import PorterStemmer
porter= PorterStemmer()
porter.stem('happiness')
```

```python
from nltk.stem import LancasterStemmer
porter = LancasterStemmer()
porter.stem('happiness')
```

```python
from nltk.stem import RegexpStemmer
porter = RegexpStemmer('ing$')
porter.stem('singing')
```

```python
from nltk.stem import SnowballStemmer
porter = SnowballStemmer('french')
porter.stem('avoir') #cling ring
```

```python
porter=PorterStemmer()
text="The figures of a horse and rider came slowly through the edding mist"
text=[porter.stem(token) for token in text.split(" ")]
print(text)
```

```python
print(" ".join(text))
```

#### Lemmatizer

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
print(lemma.lemmatize("cats"))
```

```python
print(lemma.lemmatize("am",pos='v'))
```

```python
print(lemma.lemmatize("better",pos='a'))
```
